##a).

***R/***

$$y=0$$
Por lo tanto esta secuencia **Converge**.

##b).
$$a_n = \pi^n, \quad n = 0, 1, 2, ...$$

***R/***
<br>Al ser una secuencia de tipo polinómica, es decir, al no contar con Asintota Horizontal. **No tiene Límite** Por lo tanto, **Diverge**

##c).
$$b_n = \frac{3^{n-1}}{2n + 6}, \quad n = 1, 2, 3, ...$$

***R/***
<br>Como el exponente del numerador crece exponencialmente segun la sucesión, quiere decir que su grado es mayor que al del denominador. Por lo tanto **No tiene Limite** Al no poseer Asintota H. Por lo tanto **Diverge**

##d).
$$c_n = \cos(\pi n), \quad n = 0, 1, 2, ...$$

***R/***
<br>La secuencia al ser de tipo polinómica, no posee Asintota Horizontal, lo que quiere decir que **No tiene Límite** y **Diverge**.

##e).
$$a_n = \frac{3n^4 - 6n^3 + 1}{5n^3 + n^2 + 2}, \quad n = 1, 2, 3, ...$$

***R/***
<br>En este caso como el exponente del numerador $$3n^4$$ es mayor que el del denominador $$5n^3$$ La secuencia no cuenta con Asintota Horizontal, lo que quiere decir que **No tiene Límite** y **Diverge**.

##f).
$$b_n = \frac{2n^5 - 3n^2 + 23}{7n^5 + 13n^4 - 12}, \quad n = 1, 2, 3, ...$$

***R/***
<br>En este caso al poseer el mismo de grado de exponente tanto en el numerador como en el denominador $$n^5$$ Obtenemos como **Límite** $$y= \frac{2}{7}$$ Dando como conclusión que **Converge**.

##g).
$$c_n = \frac{45 - 16n^2}{13 + 5n + 6n^3}, \quad n = 1, 2, 3, ...$$

***R/***
<br>En este caso como el grado del denominador $$n^3$$ es mayor que el del numerador $$n^2$$ Obtenemos como **Límite** $$y=0$$ Por lo tanto, la secuencia **Converge**

##h).
$$b_n = \frac{3n + 1}{\sqrt{4n^2 + 1}}, \quad n = 1, 2, 3, ...$$
***R/***
<br>Al tener el mismo grado en los exponentes del numerador y denominador $$\frac{3n}{\sqrt{4n^2}}= \frac{3n}{2n}$$ Obtenemos que el **Límite es:** $$y=\frac{3}{2}$$ Por lo tanto la secuencia **Converge**.

##i).

***R/***

##j).

```python
import math
import matplotlib.pyplot as plt

def calculate_an(n):
  return (1 + 1/n)**n

a1 = calculate_an(1)
a2 = calculate_an(2)
a3 = calculate_an(3)
a4 = calculate_an(4)
print(f"a1 = {a1}")
print(f"a2 = {a2}")
print(f"a3 = {a3}")
print(f"a4 = {a4}")
```

###b).

```python
an_values = [calculate_an(n) for n in range(1, 201)]
```

###c).

```python
n_values = list(range(1, 201))
plt.plot(n_values, an_values, label="a_n")
plt.axhline(y=math.e, color='r', linestyle='-', label="y = e")
plt.xlabel("n")
plt.ylabel("a_n")
plt.title("Sequencia a_n = (1 + 1/n)^n")
plt.legend()
plt.show()
```

###d).

***El límite de a_n cuando n se acerca al infinito es e.***

###e).

```python
for i in range(1, 201):
  if calculate_an(i) > 2.7:
    print(f"El valor más pequeño de n para el cual a_n > 2.7 is: {i}")
    break
```

###f).

```python
for i in range(1, 201):
  if abs(calculate_an(i) - math.e) < 0.01:
    print(f"El primer valor de n para el cual |a_n - e| < 0.01 es: {i}")
    break
```

```python
import matplotlib.pyplot as plt

a_0 = 1.01
b_0 = 0.99


n_values = list(range(201))
a_values = [a_0**n for n in n_values]
b_values = [b_0**n for n in n_values]


plt.figure(figsize=(10, 6))
plt.plot(n_values, a_values, label="a_n = 1.01^n", color='blue')
plt.plot(n_values, b_values, label="b_n = 0.99^n", color='red')


plt.xlabel("n")
plt.ylabel("Valor de las secuencias")
plt.title("Comparación de las secuencias a_n y b_n para n = 0, 1, 2, ..., 200")
plt.legend()
plt.grid(True)

plt.show()
```

#Analisis

-La secuencia $$a_n = 1.01^n$$ crece de manera exponencial a medida que $$n$$ aumenta, alejándose del eje horizontal.

-la secuencia $$b_n = 0.99^n$$ , en cambio, decrece exponencialmente y se aproxima a cero a medida que
$$n$$aumenta

#**7.**

```python
import numpy as np
import matplotlib.pyplot as plt


def a_n(n):
    return (10**n) / np.math.factorial(n)


n_values = list(range(1, 101))
a_values = [a_n(n) for n in n_values]

plt.figure(figsize=(10, 6))
plt.plot(n_values, a_values, label=r"$a_n = \frac{10^n}{n!}$", color='blue', marker='o', markersize=4)


plt.xlabel("n")
plt.ylabel("Valor de $a_n$")
plt.title("Gráfico de $a_n = \\frac{10^n}{n!}$ para n = 1, 2, 3, ..., 100")
plt.legend()
plt.grid(True)


plt.show()
```

#**8.**

```python
import matplotlib.pyplot as plt

def secuencia_1(n):

  if n == 1:
    return 10
  else:
    return secuencia_1(n-1) / 3

n_valores = range(1, 51)
seq1_valores = [secuencia_1(n) for n in n_valores]


plt.plot(n_valores, seq1_valores, 'o-', label="Secuencia 1")
plt.title("Secuencia 1: a sub n = 1/3 * a sub (n-1)")
plt.xlabel("n")
plt.ylabel("a_n")
plt.legend()
plt.grid(True)
plt.show()
```

# Análisis:
Al observar la gráfica, podemos ver que los términos de la secuencia disminuyen rápidamente y se acercan a cero, lo cual indica que la secuencia converge a 0.

```python
import matplotlib.pyplot as plt

def secuencia_2(n):

  if n == 1:
    return 2
  else:
    return 2 * secuencia_2(n-1)


n_valores = range(1, 51)
seq2_valores = [secuencia_2(n) for n in n_valores]

plt.plot(n_valores, seq2_valores, 'o-', label="Secuencia 2")
plt.title("Secuencia 2: a sub n = 2 * a sub (n-1)")
plt.xlabel("n")
plt.ylabel("a_n")
plt.legend()
plt.grid(True)
plt.show()
```

# Análisis:
Al observar la gráfica, podemos ver que los términos de la secuencia aumentan exponencialmente lo cual indica que la secuencia diverge hacia infinito.
